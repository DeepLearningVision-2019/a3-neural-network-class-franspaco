### Import the needed Packages

```python
import numpy as np
import matplotlib.pyplot as plt

# Needed for the mnist data
from keras.datasets import mnist
from keras.utils import to_categorical 
```

```python
import itertools

def pairwise(iterable):
    "s -> (s0,s1), (s1,s2), (s2, s3), ..."
    a, b = itertools.tee(iterable)
    next(b, None)
    return zip(a, b)
```

### Define the class

```python
class NeuralNetwork:

    def __init__(self, architecture, alpha):
        '''
            layers: List of integers which represents the architecture of the network.
            alpha:  Learning rate.
        '''
        if len(architecture) < 2:
            raise Exception("Doesn't you need at least an input and output layer.")
        self.alpha = alpha
        self.layers = architecture
        
        self.outputs = [np.array(l_size) for l_size in architecture]
        
        self.weights = [np.random.rand(prev, nxt) for prev, nxt in pairwise(architecture)]
        self.weights.insert(0,0)
        
        self.bs = [np.random.rand(1, nxt) for prev, nxt in pairwise(architecture)]
        self.bs.insert(0,0)
            
    def __repr__(self): 
        # construct and return a string that represents the network 
        # architecture 
        return "NeuralNetwork: {}".format( "-".join(str(l) for l in self.layers))

    def softmax(self, X):  
        # applies the softmax function to a set of values
        
        expX = np.exp(X)
        return expX / expX.sum(axis=1, keepdims=True)
        
    def sigmoid(self, x):
        # the sigmoid for a given input value
        return 1.0 / (1.0 + np.exp(-x))
    
    def sigmoid_deriv(self, x):
        # the derivative of the sigmoid
        #return self.sigmoid(x)*(1-self.sigmoid(x))
        return x * (1 - x)
    
    def predict(self, inputs):
        # TODO: Define the predict function
        self.outputs[0] = inputs
        for i in range(1, len(self.layers)-1):
            self.outputs[i] = self.sigmoid(self.outputs[i-1]@self.weights[i] + self.bs[i])
        self.outputs[-1] = np.dot(self.outputs[-2], self.weights[-1]) + self.bs[-1]
        return self.softmax(self.outputs[-1])
        
    def train(self, inputs, labels, epochs = 1000, displayUpdate = 100):
        # TODO: Define the training step for the network. It should include the forward and back propagation
        # steps, the updating of the weights, and it should print the error every 'displayUpdate' epochs
        # It must return the errors so that they can be displayed with matplotlib
        m = inputs.shape[0]
        m_inv = 1/m
        er_delta = [ np.zeros((m, nxt)) for prev, nxt in pairwise(self.layers)]
        er_delta.insert(0,0)
        self.outputs[0] = inputs
        
        err_list = []
        for epoch in range(epochs):
            # Forwards
            
            # Hidden Layers
            for i in range(1, len(self.layers)-1):
                self.outputs[i] = self.sigmoid(np.dot(self.outputs[i-1], self.weights[i]) + self.bs[i])
            
            # Output Layer
            self.outputs[-1] = self.softmax(np.dot(self.outputs[-2], self.weights[-1]) + self.bs[-1])
            
            # End of forwards
            
            
            # Backwards
            
            # Output Layer
            
            er_delta[-1] = labels - self.outputs[-1]
            err_list.append(np.mean(np.abs(er_delta[-1])))
            er_delta[-1] = er_delta[-1] * self.sigmoid_deriv(self.outputs[-1])
            self.weights[-1] += m_inv * self.alpha * np.dot(self.outputs[-2].T, er_delta[-1])
            self.bs[-1]      += m_inv * self.alpha * np.sum(er_delta[-1], axis=0, keepdims=True)
            
            # Hidden Layers
            for i in reversed(range(1, len(self.layers)-1)):
                er_delta[i] = np.dot(er_delta[i+1], self.weights[i+1].T)
                er_delta[i] = er_delta[i] * self.sigmoid_deriv(self.outputs[i])
                self.weights[i] += m_inv * self.alpha * np.dot(self.outputs[i-1].T, er_delta[i])
                self.bs[i]      += m_inv * self.alpha * np.sum(er_delta[i], axis=0, keepdims=True)
                
            if epoch%displayUpdate == 0:
                print(f"Error [{epoch}]: {err_list[-1]}", end = "\r")
        print("\n")
        return err_list
    
    def evaluate(self, inputs, labels):
        
        mislabeled = []
        
        for indx, (inpt, lbl) in enumerate(zip(inputs, labels)):
            val = np.argmax(self.predict(inpt))
            real = np.argmax(lbl)
            if val != real:
                mislabeled.append((indx, val, real))
        
        accuracy = 1 - len(mislabeled)/len(inputs)
        
        print(f"Accuracy: {accuracy*100}%")
        
        return mislabeled
            
```

```python
a = NeuralNetwork([2,2],0.1)
print(a)
i = np.array([[1,1]])
l = np.array([[0,1]])
print(a.predict(i))
print("Training")
errors = a.train(i,l, 10000)
print("End-Training")
plt.plot(errors)
print(a.predict(i))
```

Train the network to classify hand drawn digits.

```python
# Load the train and test data from the mnist data set
(train_images_all, train_labels_all), (test_images_all, test_labels_all) = mnist.load_data()

# Plot a sample data point
plt.title("Label: " + str(train_labels_all[0]))
plt.imshow(train_images_all[0], cmap="gray")
```

```python
# Standardize the data
size = 50000

# Flatten the images
train_images = train_images_all.reshape((60000, 28 * 28))
# turn values from 0-255 to 0-1
train_images = train_images.astype('float32') / 255 
train_images = train_images[:size]

test_images = test_images_all.reshape((10000, 28 * 28)) 
test_images = test_images.astype('float32') / 255

# Create one hot encoding for the labels
train_labels = to_categorical(train_labels_all)
train_labels = train_labels[:size]
test_labels = to_categorical(test_labels_all)

print(f"Training on {size} elements. (Size: {train_images.shape})")
print(np.max(train_images[0]))
print(np.min(train_images[0]))
print(np.max(test_images[0]))
print(np.min(test_images[0]))
```

```python
# TODO: Test the class with the mnist data. Test the training of the network with the test_images data, and 
# record the accuracy of the classification.
c = NeuralNetwork([ 784, 10], 0.5)
print(c)
errors = c.train(train_images, train_labels, 200, 10)
plt.plot(errors)
```

```python
mislabled = c.evaluate(test_images, test_labels)
```

```python
import random

error_samples  = random.sample(mislabled, k=9)


f, plots = plt.subplots((len(error_samples)+3-1)//3, 3, figsize=(20,10))
plots = [plot for sublist in plots for plot in sublist]

for (im_indx, pred, real), plot in zip(error_samples, plots):
    plot.set_title(f"Pred: {pred} - Real: {real}")
    plot.imshow(test_images[im_indx].reshape((28,28)), cmap="gray")
```
